```python
import pandas as pd
df = pd.read_csv('EstimatedResidentPopulationByStateAndGender.csv')
print("A1.1")
popu_list=[]
for i in range(df.shape[1]):
    if ('Persons' in df.columns[i]) & (df.columns[i] != 'Estimated Resident Population ;  Persons ;  Australia ;'):
        popu_list.append(i)
podf = df[df.columns[popu_list]]
podf.rename(columns=lambda x: x.split(';')[2].strip(),inplace=True)
podf
```

```python
import matplotlib.pyplot as plt
print('A1.2')
plt.plot(podf.Victoria)
plt.plot(podf['New South Wales'])
plt.plot(podf.Queensland)
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Population for three states over time')
plt.legend()
plt.show()
print('a. The population for these three states had incresed over time')
print('b. They all show the linear increase trend')
```

```python
from scipy.stats import linregress
print('A1.3')
index = range(1,40)
slope, intercept, r_value, p_value, std_err = linregress(index,podf['Victoria'])
line = [slope*xi + intercept for xi in index]
plt.scatter(index, podf['Victoria'])
plt.plot(index,line,'r-', linewidth=3)
plt.show()
plt.xlabel('Time')
plt.ylabel('Population')
plt.title('Population for three states over time')
print('The Linear equation is y =' , slope ,'x +',intercept )
print('the index for 1/9/15,1/12/15, 1/12/16 and 1/12/17 are 40, 41, 45, 49')
print('so the prediction for population of 1/12/15 is')
print(slope ,'*40' ,'+',intercept ,'=' ,slope*40+intercept)
print('so the prediction for population of 1/12/15 is')
print(slope ,'*41' ,'+',intercept ,'=' ,slope*41+intercept)
print('so the prediction for population of 1/12/16 is')
print(slope ,'*45' ,'+',intercept ,'=' ,slope*45+intercept)
print('so the prediction for population of 1/12/17 is')
print(slope ,'*49' ,'+',intercept ,'=' ,slope*49+intercept)
```
